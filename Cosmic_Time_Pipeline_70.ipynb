Setting up functions for computing p-value in KS test 

```python

def QNew(x):
    term=9999 #initialize as big number
    sum=0 #initialize as zero
    j=1 #summation index 
    while np.fabs(2*np.exp(-2*(j+1)*(j+1)*x*x)*((-1)**(j)))>(10**(-9)): #while abs value of first omitted term > 10^-9
        term=2*np.exp(-2*j*j*x*x)*((-1)**(j-1))
        sum=sum+term
        j=j+1
    return sum 

#This ensures that the error in the p-value is less than 10^-9

def pvalue(cdf1, cdf2, n1, n2):
    cdf1=np.array(cdf1)
    cdf2=np.array(cdf2)
    N=n1*n2/(n1+n2)
    D = np.max(np.abs(cdf1 - cdf2))
    pvalue=QNew((np.sqrt(N)+0.12+0.11/np.sqrt(N))*D)
    return pvalue, D
```

Generating 10,000 Empirical CDFs for CDM

```python
fig, ax_bottom = plt.subplots()
ax_bottom.set_xlabel("Cosmic Time (Gyr)")
ax_bottom.set_ylabel("Cumulative Probability")
ax_bottom.grid(True)
pdf_cdm=np.zeros((Number_Draws,len(x_vals)))
cdf_cdm=np.zeros((Number_Draws,len(x_vals)))
runs=0
while runs < Number_Draws:
    Cosmic_Times_CDM_Sampled=[]
    for element in range(1,Number_Mergers_Seen_LISA+1):
        Random_val=np.random.uniform(0,1)
        Cosmic_Times_CDM_Sampled.append(inverted_CDF_CDM(Random_val)) #inverse transform sampling 

    Cosmic_Times_CDM_Sampled=np.array(np.sort(Cosmic_Times_CDM_Sampled)) #sort in ascending order 
    CDF_CDM_Sampled = np.linspace(1/len(Cosmic_Times_CDM_Sampled),1,len(Cosmic_Times_CDM_Sampled)) #make cdf, for comparison cdfs made with kde 
    
    Times_Interpolation=np.linspace(np.min(Cosmic_Times_CDM_Sampled), np.max(Cosmic_Times_CDM_Sampled),1000)
    stepwise = interp1d(Cosmic_Times_CDM_Sampled, CDF_CDM_Sampled, kind='zero') #step-wise continuous, for comparison to smoothed curves from kde 

    kde=gaussian_kde(Cosmic_Times_CDM_Sampled, 0.15)
    pdf_cdm_vals = kde.evaluate(x_vals) #generate pdf 
    pdf_cdm[runs][:]=pdf_cdm_vals
    cdf_cdm_vals = integrate.cumulative_trapezoid(pdf_cdm_vals, x_vals, initial=0) #integrate pdf to get cdf 
    scale=np.max(np.double(cdf_cdm_vals))
    cdf_cdm_vals=np.double(cdf_cdm_vals/scale) #Normalize so it = 1 at end 
    cdf_cdm[runs][:]=cdf_cdm_vals #store in an array 
    ax_bottom.plot(Times_Interpolation, stepwise(Times_Interpolation), color='k', linewidth=3)
    ax_bottom.plot(x_vals,cdf_cdm_vals, color='r', linewidth=3)
    ax_bottom.scatter(Cosmic_Times_CDM_Sampled, CDF_CDM_Sampled, color='k')
    runs=runs+1
plt.show()
print(np.shape(cdf_cdm))
```
